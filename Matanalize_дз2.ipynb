**Задача 1**. (11 баллов)

В известном магазине техники представлено очень много экземпляров различной бытовой техники двух видов: с возможностью подключения к системе "Умный дом" и без такой возможности. Причем, известно, что эти два вида техники представлены в пропорции 1 : 3 соответственно. Случайным образом выбрали 7 экземпляров техники для розыгрыша среди участниц конкурса "Мисс Рукоделье - 2024".  Какова вероятность, что среди них не более двух экземпляров с возможностью подключения к системе "Умный дом"?

Подробно распишите решение и обоснуйте выбор распределения для моделирования сюжета задачи. Ответ округлите до трёх знаков.


Для решения задачи используем биномиальное распределение, так как:

Фиксированное число испытаний: 7 экземпляров техники.

Два исхода: техника либо поддерживает "Умный дом", либо нет.

Постоянная вероятность успеха: вероятность того, что техника поддерживает "Умный дом", равна p=1/4, так как пропорция 1:3.



```python
import scipy as sp

# Параметры биномиального распределения
n = 7  # количество испытаний
p = 1/4  # вероятность успеха (техника с поддержкой "Умного дома")

# Вычисляем вероятность P(k <= 2)
probability = sp.stats.binom.cdf(2, n, p)

# Округляем результат до трёх знаков
probability_rounded = round(probability, 3)

probability_rounded
```

**Задача 2** (30 баллов)

Дан закон распределения количества публикаций в которой группе телеграм-каналов:

$$
\begin{array}{|c|c|c|c|c|c|c|}
\hline X \text { (число публикаций) } & 0 & 1 & 2 & 3 & 4 & 5 \\
\hline \mathrm{P} \text { (вероятность) } & ? & 0,3 & 0,3 & 0,2 & 0,1 & 0,1 \\
\hline
\end{array}
$$

Найдите пропущенную вероятность для X = 0, математическое ожидание и дисперсию. Решите задачу "вручную" и с помощью python.

1. Найдём пропущенную вероятность для X=0:
Сумма всех вероятностей должна быть равна 1. Обозначим пропущенную вероятность как P(0).

$$
P(0) + 0.3 + 0.3 +0.2 +0.1 + 0.1 = 1
$$

Таким образом, пропущенная вероятность P(0) = 0

2. Найдём математическое ожидание E(X):
Математическое ожидание вычисляется по формуле:
$$
E(X)=∑xi⋅P(xi)
$$

Подставляем значения:

$$
E(X)=0*0+1*0.3+2*0.3+3*0.2+4*0.1+5*0.1
E(X)= 2.4
$$

3. Найдём дисперсию D(X):
Дисперсия вычисляется по формуле:
$$
D(X)=E(X**2)−[E(X)]**2
$$

сначала найдем E(X**2)
E(X**2) = 7,4
Теперь найдём дисперсию:
$$
D(X)=7.4 - (2,4)**2
D(X)= 1.64
$$

```python
import numpy as np

# Данные
X = np.array([0, 1, 2, 3, 4, 5])
P = np.array([0, 0.3, 0.3, 0.2, 0.1, 0.1])

# Математическое ожидание
E_X = np.sum(X * P)

# Дисперсия
E_X2 = np.sum(X**2 * P)
D_X = E_X2 - E_X**2

print(f"Пропущенная вероятность P(0): {P[0]}")
print(f"Математическое ожидание E(X): {E_X}")
print(f"Дисперсия D(X): {round(D_X, 2)}")
```

**Задача 3** (25 баллов)



Компания приобрела 10 000 новых ноутбуков. Вероятность поломки одного ноутбука в течение месяца равна 0,0002. Предполагается, что поломки отдельных ноутбуков происходят независимо друг от друга.

**Вопрос 1**. Найдите вероятность того, что за месяц сломается ровно 1 ноутбук.
Решение объясните. (5 баллов)

**Вопрос 2**. Компания планирует ввести программу расширенной гарантии, по которой она будет бесплатно ремонтировать или заменять ноутбуки, сломавшиеся в течение месяца. Если общее количество поломок за месяц превысит 5 ноутбуков, компания понесет значительные дополнительные расходы.

а) Найдите вероятность того, что за месяц сломается более 5 ноутбуков.Решение объясните (10 баллов)

б) Если стоимость ремонта или замены одного ноутбука составляет 500 долларов, вычислите математическое ожидание общих расходов компании на поломки за один месяц. (10 баллов)

1. Для решения задачи используем биномиальное распределение.

10 000 * 0.0002 * 0.1353 ~ 0.2706

Таким образом, Вероятность того, что за месяц сломается ровно 1 ноутбук, составляет 0,2706.

2.
а.  Используем формулу Пуассона для аппроксимации биномиального распределения, так как n велико, а p мало.
б.Математическое ожидание общих расходов компании на поломки за один месяц составляет 1000 долларов.

```python
# Параметры задачи
n = 10000  # количество ноутбуков
p = 0.0002  # вероятность поломки одного ноутбука
cost_per_laptop = 500  # стоимость ремонта одного ноутбука

# Вопрос 1: Вероятность того, что сломается ровно 1 ноутбук
prob_1 = sp.stats.binom.pmf(k=1, n=n, p=p)
print(f"Вопрос 1: Вероятность того, что сломается ровно 1 ноутбук: {prob_1:.4f}")

# Вопрос 2а: Вероятность того, что сломается более 5 ноутбуков
lambda_ = n * p  # параметр для пуассоновского распределения
prob_more_than_5 = 1 - sp.stats.poisson.cdf(k=5, mu=lambda_)
print(f"Вопрос 2а: Вероятность того, что сломается более 5 ноутбуков: {prob_more_than_5:.4f}")

# Вопрос 2б: Математическое ожидание общих расходов
expected_expenses = lambda_ * cost_per_laptop
print(f"Вопрос 2б: Математическое ожидание общих расходов: {expected_expenses} долларов")
```
