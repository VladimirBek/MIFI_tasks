$

\end{pmatrix}
$

$

Вычисляем определитель матрицы:

$\det(B) = 1 \cdot (2 \cdot 2 - 1 \cdot (-4)) - 2 \cdot (0 \cdot 2 - 1 \cdot 2) + 3 \cdot (0 \cdot (-4) - 2 \cdot 2) = 0$

Ответ: Так как определитель подматрицы равен 0, векторы линейно зависимы при любом значении параметра "а".

```python
# Преобразуем DataFrame в матрицу
matrix = df.values

# Находим ранг матрицы
rank = np.linalg.matrix_rank(matrix)

# Получаем колличество стобцов

num_cols = matrix.shape[1]



print(f"Ранг матрицы: {rank}")
print(f"Количество столбцов: {num_cols}")
if rank < num_cols:
    print("Есть линейная зависимость между некоторыми столбцами")
    corr_matrix = df.corr()

    # Находим пары столбцов с высокой корреляцией (больше 0.9, но меньше 1.0)
    high_corr_pairs = corr_matrix[(corr_matrix > 0.9) & (corr_matrix < 1.0)]

    # Убираем пустые строки и столбцы для удобства чтения
    high_corr_pairs_cleaned = high_corr_pairs.dropna(how='all').dropna(axis=1, how='all')

    print("\nПары столбцов с высокой корреляцией:")
    if not high_corr_pairs_cleaned.empty:
        print(high_corr_pairs_cleaned)
    else:
        print("Нет пар столбцов с корреляцией выше 0.9.")
else:
    print("\nЛинейная зависимость между столбцами отсутствует.")
    
```

Задание 10.

Разделите данные на предикторы и целевую переменную. Обучите модель с помощью классической линейной регрессии. Линейную регрессию необходимо реализовать без готовых библиотек.

Оцените качество модели с помощью хотя бы двух метрик качества.

Распишите решение с помощью кода на языке Python и напишите текстом интерпретацию полученного вывода.

```python
x = np.array(df.drop('cnt', axis=1))
y = np.array(df['cnt'].values)

x = np.column_stack([np.ones(len(x)), x])

# вычислим OLS-оценку для коэффициентов
w_hat = np.linalg.inv(x.T@x)@x.T@y

y_pred = x @ w_hat

# Вычисление метрик
mse = mean_squared_error(y, y_pred)  # Среднеквадратичная ошибка (MSE)
mae = mean_absolute_error(y, y_pred)  # Средняя абсолютная ошибка (MAE)
r2 = r2_score(y, y_pred)  # Коэффициент детерминации (R²)

# Вывод метрик
print(f"Среднеквадратичная ошибка (MSE): {mse}")
print(f"Средняя абсолютная ошибка (MAE): {mae}")
print(f"Коэффициент детерминации (R²): {r2}")
```
