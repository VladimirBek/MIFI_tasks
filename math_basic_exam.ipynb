### Задание 12

Вероятность, что первый телефон сломан - 0.1. Вероятность того, что второй телефон сломан - 0.3. Найдите вероятность, что оба телефона работают.

Вероятность того, что первый телефон работает:

$
P(\text{первый работает}) = 1 - P(\text{первый сломан}) = 1 - 0.1 = 0.9
$

Вероятность того, что второй телефон работает:

$
P(\text{второй работает}) = 1 - P(\text{второй сломан}) = 1 - 0.3 = 0.7
$

Вероятность того, что оба телефона работают:

$
P(\text{оба работают}) = P(\text{первый работает}) \times P(\text{второй работает}) = 0.9 \times 0.7 = 0.63
$

Ответ: 0,63

### Задание 17

1. Треугольник ABC — равнобедренный, поэтому BAC = ABC , и tgABC= $\frac{7}{24}$.

$
tg(\angle ABC) = \frac{AH}{BH} = \frac{7}{24} \implies AH = 7x, \quad BH = 24x.

3. По теореме Пифагора найдем AB:

$
