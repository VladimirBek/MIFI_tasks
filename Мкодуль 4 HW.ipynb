1. Применяемая статистика и обоснование ее применения
Мы используем z-тест для проверки гипотезы о среднем значении.
Обоснование:

Генеральное стандартное отклонение известно.
Размер выборки достаточно большой, что позволяет использовать центральную предельную теорему.

2-6. Расчетное значение статистики, кричитеское значение статистики и p-value вычисляются в коде в следующей ячейке. Статичкский вывод также делается в ходе выполнения кода.

7. Содержательный вывод. 

На основании выборки из 50 коробок порошка, с уровнем значимости 5%, у инженера-технолога есть достаточно доказательств, чтобы утверждать, что истинное среднее значение веса коробок не равно 3,00 кг. Это говорит о том, что дозатор требует настройки.

```python
import scipy.stats as stats
import math
import numpy as np

mu_0 = 3.00       
x_bar = 3.005      
sigma = 0.015      
n = 50             
alpha = 0.05       


z_stat = (x_bar - mu_0) / (sigma / math.sqrt(n))
print(f"Z-статистика: {z_stat:.4f}")


z_critical = stats.norm.ppf(1 - alpha / 2) 
print(f"Критическое значение Z: ±{z_critical:.4f}")


p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))  
print(f"P-value: {p_value:.4f}")


if abs(z_stat) > z_critical:
    print("Мы отвергаем нулевую гипотезу: истинное среднее не равно 3.00 кг.")
else:
    print("Мы не отвергаем нулевую гипотезу: недостаточно доказательств, что истинное среднее отличается от 3.00 кг.")

if p_value < alpha:
    print("Мы отвергаем нулевую гипотезу на основе P-value: истинное среднее не равно 3.00 кг.")
else:
    print("Мы не отвергаем нулевую гипотезу на основе P-value.")

# Содержательный вывод
if abs(z_stat) > z_critical or p_value < alpha:
    print("\nВывод: Машина требует настройки.")
else:
    print("\nВывод: Машина не требует настройки.")
```

7. На уровне значимости 5% имеются статистически значимые различия между средними прочностными характеристиками двух веществ. Это означает, что прочностные характеристики веществ отличаются, и производитель должен учитывать это при принятии решения о замене стальных компонентов алюминиевыми.


```python
x1, s1, n1 = 90, 2.3, 15 
x2, s2, n2 = 88, 2.2, 10  
alpha = 0.05 

t_stat, p_value_x2 = stats.ttest_ind_from_stats(mean1=x1, std1=s1, nobs1=n1, mean2=x2, std2=s2, nobs2=n2)

p_value = p_value_x2/2
df = n1 + n2 - 2
t_critical = stats.t.ppf(1 - alpha, df)


print(f"T-статистика: {t_stat:.4f}")
print(f"Критическое значение T: ±{t_critical:.4f}")
print(f"P-value: {p_value:.4f}")

if abs(t_stat) > t_critical:
    print("Мы отвергаем нулевую гипотезу: есть разница между средними прочностными характеристиками.")
else:
    print("Мы не отвергаем нулевую гипотезу: недостаточно доказательств различий между веществами.")

if p_value < alpha:
    print("Мы отвергаем нулевую гипотезу на основе P-value.")
else:
    print("Мы не отвергаем нулевую гипотезу на основе P-value.")
```

1. Мы имеем данные, где каждый тест проводится для двух схем (старой и новой). Это говорит о наличии связанных выборок, и поэтому применяется парный t-тест. Парный t-тест используется для проверки гипотезы о равенстве средних двух связанных выборок.

2-6. Описаны в коде в ячейке ниже


7. На уровне значимости 5% можно сделать вывод, что средние токи, пропускаемые старой и новой конструкцией, отличаются. Поэтому производителю необходимо учитывать, что новая конструкция не пропускает тот же средний ток, и потребуется дополнительная проверка перед внедрением новой конструкции.

```python
old_currents = np.array([80.1, 82.3, 84.1, 82.6, 85.3, 81.3, 83.2, 81.7, 82.2, 81.4])
new_currents = np.array([80.7, 81.3, 84.6, 81.7, 86.3, 84.3, 83.7, 84.7, 82.8, 84.4, 85.2, 84.9])
t_stat, p_value = stats.ttest_ind(old_currents, new_currents)
t_critical = stats.t.ppf(1 - alpha / 2, old_currents.size + new_currents.size - 2)

print(f"T-статистика: {t_stat:.4f}")
print(f"Критическое значение T: ±{t_critical:.4f}")
print(f"P-value: {p_value:.4f}")

if abs(t_stat) > t_critical:
    print("Мы отвергаем нулевую гипотезу: средние токи различны.")
else:
    print("Мы не отвергаем нулевую гипотезу: средние токи одинаковы.")

if p_value < alpha:
    print("Мы отвергаем нулевую гипотезу на основе P-value.")
else:
    print("Мы не отвергаем нулевую гипотезу на основе P-value.")
```
